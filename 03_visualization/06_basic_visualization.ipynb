```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Link: https://www.data-blogger.com/2017/11/15/python-matplotlib-pyplot-a-perfect-combination/
data_path = "home-credit-default-risk/application_train.csv"
pdf_data = pd.read_csv(data_path)
pdf_data.head()
```

# Matplotlib

## Line plot

```python
series_data = pdf_data["AMT_INCOME_TOTAL"][:10]
x = list(series_data.index)
y = list(series_data.values)

plt.plot(x, y)
plt.show()
```

## Bar plot

```python
series_data = pdf_data["NAME_CONTRACT_TYPE"].value_counts()
x = list(series_data.index)
y = list(series_data.values)

plt.bar(x, y)
plt.show()
```

## Historgram

```python
series_data = pdf_data["AMT_CREDIT"]
x = list(series_data.values)

plt.hist(x, bins=20)
plt.show()
```

## Box plot

```python
series_data = pdf_data[pdf_data["TARGET"] == 0]["AMT_CREDIT"]
x1 = list(series_data.values)

series_data = pdf_data[pdf_data["TARGET"] == 1]["AMT_CREDIT"]
x2 = list(series_data.values)

plt.boxplot([x1, x2])
plt.show()
```

## Scatter plot

```python
series_data = pdf_data["AMT_CREDIT"]
x1 = list(series_data.values)

series_data = pdf_data["AMT_ANNUITY"]
x2 = list(series_data.values)

plt.scatter(x1, x2)
plt.show()
```

# Pandas

## Line plot

```python
series_data = pdf_data["AMT_INCOME_TOTAL"][:10]
series_data.plot()
plt.show()
```

## Bar plot

```python
series_data = pdf_data["NAME_CONTRACT_TYPE"].value_counts()
series_data.plot.bar()
plt.show()
```

## Historgram

```python
series_data = pdf_data["AMT_CREDIT"]
series_data.hist(bins=20)
plt.show()
```

## Box plot

```python
pdf_data.boxplot(column="AMT_CREDIT", by="TARGET")
plt.show()
```

## Scatter plot

```python
pdf_data.plot.scatter(x="AMT_CREDIT", y="AMT_ANNUITY")
plt.show()
```

# Matplotlib subplots

![Matplotlib properties](subdivision.png "Matplotlib properties")

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))

# plot at axes(0, 0)
pdf_data["AMT_INCOME_TOTAL"][:10].plot(ax=axes[0][0])

# plot at axes(0, 1)
pdf_data["NAME_CONTRACT_TYPE"].value_counts().plot.bar(ax=axes[0][1])

# plot at axes(1, 0)
pdf_data["AMT_CREDIT"].hist(bins=20, ax=axes[1][0])

# plot at axes(1, 1)
pdf_data.plot.scatter(x="AMT_CREDIT", y="AMT_ANNUITY", ax=axes[1][1])

plt.show()
```
